```python
Write a Python program to find a target values in a list using linear search with following steps:

a. Initialize the fist to store the input elements.

b. Initialize found-False.

C. Enter the item to be searched (match_item).

d. For each element in the list

1. if match item = value

a. return match item's position.

e. If the match item is not in the list, display an error message that the item is not found in the list
```

```python
def linear_search(input_list, match_item):
    found = False

    for i, value in enumerate(input_list):
        if match_item == value:
            found = True
            return f"{match_item} found at position {i + 1}"

    if not found:
        return f"{match_item} is not found in the list"

# Initialize the input list
input_list = [10, 25, 5, 30, 15, 50, 20]

# Input the item to be searched
match_item = int(input("Enter the item to be searched: "))

result = linear_search(input_list, match_item)
print(result)
```

```python
Write a Python program to implement binary search to find the target values from the list: 

a.create a sepearate function to do binary search

b. Get the number of inputs from the user.

C. Store the inputs individually  in a list.

d. In binary search function af sort the list order to  start the search from middle of t list.

e. Compare the middle element

elements to search target element.

f. If greater, move to right of list or else move to another side of the list.

g. Print the result along with the position of the element.
```

```python
def binary_search(input_list, target):
    # Sort the list in ascending order (binary search requires a sorted list)
    input_list.sort()
    
    left = 0
    right = len(input_list) - 1

    while left <= right:
        mid = (left + right) // 2
        mid_value = input_list[mid]

        if mid_value == target:
            return f"{target} found at position {mid + 1}"
        elif mid_value < target:
            left = mid + 1
        else:
            right = mid - 1

    return f"{target} is not found in the list"

# Get the number of inputs from the user
num_inputs = int(input("Enter the number of inputs: "))

# Store the inputs individually in a list
input_list = []
for i in range(num_inputs):
    input_value = int(input(f"Enter input {i + 1}: "))
    input_list.append(input_value)

# Input the target element to search
target_element = int(input("Enter the target element to search: "))

# Call the binary_search function
result = binary_search(input_list, target_element)
print(result)
```

```python
3)Write a Python program for sorting a list of elements using selection sort algorithm:

a. Assume two lists: So list- Initially empty and Unsorted List-Given input list. 
    b. In the first iteration, find the mallest element in the unsorted list and place it in the sorted

list.


C. In the second iteration. find the lement in the unsorted list and place it in the correct position by comparing with the element in the sorted list.

d. In the third iteration, again find the smallest element in the unsorted list and place it in the correct position by comparing with the elements in the sorted list.

e. This process continues till the unsorted list becomes empty.

f. Display the sorted list
```

```python
def selection_sort(input_list):
    sorted_list = []

    while input_list:
        # Find the smallest element in the unsorted list
        min_element = min(input_list)

        # Remove the smallest element from the unsorted list
        input_list.remove(min_element)

        # Place the smallest element in the sorted list
        sorted_list.append(min_element)

    return sorted_list

# Input the list of elements
input_list = [64, 25, 12, 22, 11]

# Call the selection_sort function to sort the list
sorted_list = selection_sort(input_list)

# Display the sorted list
print("Sorted list:", sorted_list)
```

```python
def insertion_sort(input_list):
    for i in range(1, len(input_list)):
        current_element = input_list[i]
        j = i - 1

        while j >= 0 and current_element < input_list[j]:
            input_list[j + 1] = input_list[j]
            j -= 1

        input_list[j + 1] = current_element

# Get the number of elements from the user
num_elements = int(input("Enter the number of elements: "))

# Initialize the input list
input_list = []

# Get input elements from the user
for i in range(num_elements):
    element = int(input(f"Enter element {i + 1}: "))
    input_list.append(element)

# Call the insertion_sort function to sort the list
insertion_sort(input_list)

# Display the sorted list
print("Sorted List:", input_list)
```

```python
5)Write a Python program that 3
merge sort on a list of numbers:

a. Divide: If the given array has zero or one element, return.

1. Otherwise

ii. Divide the input list in to two halves each containing half of the elements. i.e. left half and right half.

b. Conquer: Recursively sort the two lists (left half and right half).

a. Call the merge sort on left half.

b. Call the merge sort on right half.

C. Combine: Combine the elements back in the input list by merging the two sorted lists into a sorted sequence.
```

```python
def merge_sort(input_list):
    if len(input_list) <= 1:
        return input_list

    # Divide the input list into two halves
    mid = len(input_list) // 2
    left_half = input_list[:mid]
    right_half = input_list[mid:]

    # Recursively sort the two halves
    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    # Merge the sorted halves back into a sorted sequence
    sorted_list = merge(left_half, right_half)

    return sorted_list

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # Append any remaining elements in the left and right lists
    result.extend(left[i:])
    result.extend(right[j:])

    return result

# Input the list of numbers
input_list = [38, 27, 43, 3, 9, 82, 10]

# Call the merge_sort function to sort the list
sorted_list = merge_sort(input_list)

# Display the sorted list
print("Sorted list:", sorted_list)
```

```python
6)Write a Python script to perform the following operations on a singly linked list

a. Create a list

b. Find the smallest element from the list C. Insert an element if it is not a duplicate element

d. Display the elements in reverse order
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def find_smallest(self):
        if not self.head:
            return None

        current = self.head
        smallest = current.data
        while current:
            if current.data < smallest:
                smallest = current.data
            current = current.next

        return smallest

    def insert_if_not_duplicate(self, data):
        if not self.head:
            self.insert(data)
            return

        current = self.head
        while current:
            if current.data == data:
                return
            if current.next is None:
                break
            current = current.next
        self.insert(data)

    def display_reverse(self):
        if not self.head:
            print("List is empty")
            return

        stack = []
        current = self.head
        while current:
            stack.append(current.data)
            current = current.next

        while stack:
            print(stack.pop(), end=" -> ")
        print("None")

# Create a singly linked list
linked_list = LinkedList()

# Insert elements into the list
elements = [10, 5, 15, 20, 5, 25, 10]
for element in elements:
    linked_list.insert_if_not_duplicate(element)

# Find the smallest element
smallest = linked_list.find_smallest()
print("Smallest element:", smallest)

# Display elements in reverse order
print("Elements in reverse order:")
linked_list.display_reverse()
```

```python
7)
Write a python program to implement the various operations for Stack ADT

i.) Push ii.) Pop iii.) Display.
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return "Stack is empty"

    def display(self):
        if not self.is_empty():
            print("Stack elements:")
            for item in reversed(self.items):
                print(item)
        else:
            print("Stack is empty")

# Create a stack
stack = Stack()

while True:
    print("\nStack Operations:")
    print("1. Push")
    print("2. Pop")
    print("3. Display")
    print("4. Quit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        item = input("Enter element to push: ")
        stack.push(item)
    elif choice == 2:
        popped = stack.pop()
        print("Popped:", popped)
    elif choice == 3:
        stack.display()
    elif choice == 4:
        break
    else:
        print("Invalid choice. Please enter a valid option.")
```

```python
8) Write a python script to implement the various operations for Queue ADT i.) Insert ii.) Delete iii.) Display.
```

```python
class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def insert(self, item):
        self.items.append(item)

    def delete(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            return "Queue is empty"

    def display(self):
        if not self.is_empty():
            print("Queue elements:")
            for item in self.items:
                print(item)
        else:
            print("Queue is empty")

# Create a queue
queue = Queue()

while True:
    print("\nQueue Operations:")
    print("1. Insert")
    print("2. Delete")
    print("3. Display")
    print("4. Quit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        item = input("Enter element to insert: ")
        queue.insert(item)
    elif choice == 2:
        deleted = queue.delete()
        print("Deleted:", deleted)
    elif choice == 3:
        queue.display()
    elif choice == 4:
        break
    else:
        print("Invalid choice. Please enter a valid option.")
```

```python
9) Write a program in python to convert the following infix expression to its postfix form using push and pop operations of a Stack

a. A/B^C+D E-F*G

b. (B^2-4*A*C)^(1/2) (100
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

def infix_to_postfix(expression):
    precedence = {'^': 3, '*': 2, '/': 2, '+': 1, '-': 1, '(': 0}
    output = []
    operator_stack = Stack()

    def is_operator(token):
        return token in precedence.keys()

    for token in expression.split():
        if token.isalnum():
            output.append(token)
        elif token == '(':
            operator_stack.push(token)
        elif token == ')':
            while not operator_stack.is_empty() and operator_stack.peek() != '(':
                output.append(operator_stack.pop())
            if not operator_stack.is_empty() and operator_stack.peek() == '(':
                operator_stack.pop()
        elif is_operator(token):
            while (not operator_stack.is_empty() and
                   precedence.get(token, 0) <= precedence.get(operator_stack.peek(), 0)):
                output.append(operator_stack.pop())
            operator_stack.push(token)

    while not operator_stack.is_empty():
        output.append(operator_stack.pop())

    return ' '.join(output)

# Test cases
infix_expression_a = "A/B^C+D E-F*G"
infix_expression_b = "(B^2-4*A*C)^(1/2) (100)"

postfix_a = infix_to_postfix(infix_expression_a)
postfix_b = infix_to_postfix(infix_expression_b)

print("Postfix expression for a:", postfix_a)
print("Postfix expression for b:", postfix_b)
```
